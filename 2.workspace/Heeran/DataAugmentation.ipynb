Rotation

```
import glob 
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os 
import imutils 


angle = [0, 5, 355]


files = glob.glob("/content/gdrive/MyDrive/darknet/bin/darknet/train_2.1_rb/r1mini/" + "*.jpg")
# 파일명 
for file in files:  
   image = cv2.imread(file, cv2.IMREAD_UNCHANGED) # 흑백 이미지로 로드
   for i in angle:
     rotated = imutils.rotate(image, i)
     head_tail = os.path.split(file) 
     img_name = head_tail[1]
     print(img_name)
     print(i)
     cv2.imwrite("/content/gdrive/MyDrive/darknet/bin/darknet/50to300_flip/"+ str(i)+img_name , rotated)
```

Flip

```
angle = [0, 5, 355]

files = glob.glob("/content/gdrive/MyDrive/darknet/bin/darknet/train_2.1_rb/r1mini/" + "*.jpg")
# 파일명 
for file in files:  
   image = cv2.imread(file, cv2.IMREAD_UNCHANGED) # 흑백 이미지로 로드
   for i in angle:
     filped = cv2.flip(image, 1)
     rotated = imutils.rotate(filped, i)
     head_tail = os.path.split(file) 
     img_name = head_tail[1]
     print(img_name)
     print(i)
     cv2.imwrite("/content/gdrive/MyDrive/darknet/bin/darknet/50to300_flip/"+ "filp" + str(i)+img_name , rotated)
```

Grayscale

```
path = './'
file_list = os.listdir(path)
img_list = list()
for file in file_list:
    if file[-1] == 'g':
        img_list.append(file)
img_list

for i in img_list:
    img = cv2.imread(i, cv2.IMREAD_GRAYSCALE) # 흑백 이미지로 로드
    cv2.imwrite('./to_gray/'+ i, img) # 이미지를 저장
```

Zoom-out

```
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import glob 
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os 



files = glob.glob("./" + "*.jpg")

# load img from file
for file in files:  
   img = cv2.imread(file, cv2.IMREAD_UNCHANGED) # no color change
   data = img_to_array(img)
   samples = expand_dims(data, 0)
   datagen = ImageDataGenerator(zoom_range=[3.0,4.0])
   it = datagen.flow(samples, batch_size=1)
   i = 5
   pyplot.subplot(330 + 1 + i) 
   batch = it.next()
   image = batch[0].astype('uint8')
   pyplot.imshow(image)
   pyplot.show()

   head_tail = os.path.split(file) # file name frame1
   img_name = head_tail[1]
   print(img_name) # frame1
   cv2.imwrite("./zoom_out/"+img_name , image)
```

Brightness

```
# example of brighting image augmentation
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
# load the image

import glob 
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os 



files = glob.glob("/content/gdrive/MyDrive/darknet/bin/darknet/train_2.1_300_frames/300_frames_final/" + "*.jpg")

# load img from file
for file in files:  
   img = cv2.imread(file, cv2.IMREAD_UNCHANGED) # no color change
   data = img_to_array(img)
   samples = expand_dims(data, 0)
   datagen = ImageDataGenerator(brightness_range=[1.5,1.5])
   it = datagen.flow(samples, batch_size=1)
   i = 5
   pyplot.subplot(330 + 1 + i) 
   batch = it.next()
   image = batch[0].astype('uint8')
   pyplot.imshow(image)
   pyplot.show()

   head_tail = os.path.split(file) # file name frame1
   img_name = head_tail[1]
   print(img_name) # frame1
   cv2.imwrite("/content/gdrive/MyDrive/darknet/bin/darknet/train_2.1_300_frames/300_w_brightness/"+img_name , image)
```
